```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv('data.csv')
df.head()
```

```python
dfnew = df._get_numeric_data()
```

```python
dfnew.describe()
```

```python
df['B_Age'] = df['B_Age'].fillna(np.mean(df['B_Age']))
df['B_Height'] = df['B_Height'].fillna(np.mean(df['B_Height']))
df['R_Age'] = df['R_Age'].fillna(np.mean(df['R_Age']))
df['R_Height'] = df['R_Height'].fillna(np.mean(df['R_Height']))
```

```python
bage = df['B_Age']
rage = df['R_Age']
```

```python
plt.figure(figsize=(15, 5))
fig1 = plt.subplot(1, 2, 1)
sns.distplot(bage)

fig2 = plt.subplot(1, 2, 2) 
sns.distplot(rage)

plt.show()
```

```python
above35 =['above35' if i >= 35 else 'below35' for i in df.B_Age]
df_B = pd.DataFrame({'B_Age':above35})
df_R = pd.DataFrame({'R_Age':above35})
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.countplot(x ='B_Age', data = df_B)

plt.subplot(1, 2, 2)
sns.countplot(x ='R_Age', data = df_R)
plt.ylabel('Number of Red Fighters')
plt.title('Age of Red Fighters', color = 'red',size = 20)
```

```python
plt.figure(figsize=(14, 7))
fig = sns.kdeplot(df['B_Height'], shade=True, color='r')
fig = sns.kdeplot(df['R_Height'], shade=True, color='b')
```

```python
g = sns.FacetGrid(df, col='winby', height=3)
g.map(plt.hist, "R_Age",bins=20)
```
